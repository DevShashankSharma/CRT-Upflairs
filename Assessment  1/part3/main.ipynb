```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating the dataset
data = {
    'Student': ['John', 'Sarah', 'David', 'Emily'],
    'Math': [80, 95, 70, 85],
    'Science': [75, 85, 65, 90],
    'English': [90, 80, 75, 88],
    'History': [70, 85, 80, 92]
}

# Creating a DataFrame
df = pd.DataFrame(data)
```

# 23. Calculate the mean score of each subject using NumPy

```python
subject_means = df.iloc[:, 1:].mean().to_dict()
print("Mean Score of Each Subject:", subject_means)
```

# 24. Calculate the average score of each student

```python
df["Average Score"] = df.iloc[:, 1:].mean(axis=1)
print("\nAverage Score of Each Student:\n", df[['Student', 'Average Score']])
```

# 25. Find the student with the highest overall score

```python
df["Total Score"] = df.iloc[:, 1:5].sum(axis=1)
top_student = df.loc[df["Total Score"].idxmax()]
print("\nStudent with the Highest Overall Score:\n", top_student)
```

# 26. Scatter plot: Math vs Science scores

```python
plt.figure(figsize=(6, 4))
plt.scatter(df['Math'], df['Science'], color='blue', alpha=0.6)
plt.xlabel("Math Score")
plt.ylabel("Science Score")
plt.title("Math vs Science Scores")
plt.grid(True)
plt.show()
```

# 27. Standardize the scores using NumPy

```python
standardized_scores = (df.iloc[:, 1:5] - df.iloc[:, 1:5].mean()) / df.iloc[:, 1:5].std()
df_standardized = df[['Student']].join(standardized_scores)
print("\nStandardized Scores:\n", df_standardized)
```

# 28. Bar plot: Visualizing total scores

```python
plt.figure(figsize=(8, 5))
plt.bar(df['Student'], df['Total Score'], color=['blue', 'green', 'red', 'orange', 'purple', 'brown', 'pink', 'cyan'])
plt.xlabel("Student")
plt.ylabel("Total Score")
plt.title("Total Scores of Students")
plt.xticks(rotation=45)
plt.show()
```

# 29. Pie chart: Subject-wise average scores

```python
plt.figure(figsize=(6, 6))
plt.pie(subject_means.values(), labels=subject_means.keys(), autopct="%1.1f%%", colors=['blue', 'green', 'red', 'orange'])
plt.title("Subject-wise Average Scores")
plt.show()
```

# 30. Students with History score greater than the average History score

```python
history_avg = df['History'].mean()
above_avg_history = df[df['History'] > history_avg]
print("\nStudents with History Score Above Average:\n", above_avg_history[['Student', 'History']])
```
