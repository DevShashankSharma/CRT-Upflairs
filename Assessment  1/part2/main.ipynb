```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Creating the dataset
data = {
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 28],
    'Department': ['HR', 'IT', 'Sales', 'HR', 'IT'],
    'Salary': [50000, 70000, 45000, 52000, 80000]
}

# Creating a DataFrame
df = pd.DataFrame(data)
```

# 20. Plot a bar chart to visualize the number of employees in each department

```python
dept_counts = df['Department'].value_counts()

plt.figure(figsize=(8, 5))
dept_counts.plot(kind='bar', color=['blue', 'green', 'red'])
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.title("Number of Employees in Each Department")
plt.xticks(rotation=0)
plt.show()
```

# 21. Filter out employees with salary greater than 60,000

```python
high_salary_employees = df[df['Salary'] > 60000]
print("Employees with salary greater than 60,000:")
print(high_salary_employees)
```

# 22. Create a line graph for the salary trend by department

```python
plt.figure(figsize=(8, 5))
for department in df['Department'].unique():
    dept_data = df[df['Department'] == department]
    plt.plot(dept_data['Employee'], dept_data['Salary'], marker='o', label=department)

plt.xlabel("Employee")
plt.ylabel("Salary")
plt.title("Salary Trend by Department")
plt.xticks(rotation=45)
plt.legend()
plt.show()
```
