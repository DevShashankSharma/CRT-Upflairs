```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating the dataset
data = {
    'Employee': ['Jhon', 'Alice', 'Bob', 'Emma', 'Ryan'],
    'Age': [28, 24, 30, 26, 35],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Salary': [50000, 60000, 55000, 62000, 72000],
    'Department': ['IT', 'HR', 'Finance', 'IT', 'Finance'],
}
# Creating a DataFrame
df = pd.DataFrame(data)
```

# 17. Calculate the average salary using NumPy

```python
# Calculate the average salary using NumPy
average_salary = np.mean(df['Salary'])
print(f"Average Salary: {average_salary}")
```

# 18. Filter out employees with salary greater than 60000

```python
# Filter out employees with salary greater than 60000
high_salary_employees = df[df['Salary'] > 60000]
print("Employees with salary greater than 60000:")
print(high_salary_employees)
```

# 19. Plot department-wise average salary

```python
# Plot department-wise average salary
avg_salary_per_department = df.groupby('Department')['Salary'].mean()
plt.figure(figsize=(8, 5))
avg_salary_per_department.plot(kind='bar', color=['blue', 'green', 'red'])
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.title("Department-wise Average Salary")
plt.xticks(rotation=0)
plt.show()
```
